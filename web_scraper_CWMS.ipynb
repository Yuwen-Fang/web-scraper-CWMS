```python
# insert the date
from datetime import datetime, timedelta

start_date_str = "2022/7/1"
end_date_str = "2022/7/3"

#using string formatting to remove leading zeros from the month and day
start_date = datetime.strptime(start_date_str, "%Y/%m/%d")
end_date = datetime.strptime(end_date_str, "%Y/%m/%d")

date_list = []

#using the string format "{}/{}/{}" and pass in start_date.year, start_date.month, and start_date.day as arguments.
#appending the date string to date_list and increment start_date by one day using timedelta(days=1)
while start_date <= end_date:
    date_str = "{}/{}/{}".format(start_date.year, start_date.month, start_date.day)
    date_list.append(date_str)
    start_date += timedelta(days=1)

print(date_list)
```

```python
# 爬資料
def scrape_data(driver, NAME, ITEM):
    # 選名稱
    driver.find_element(By.XPATH, '//*[@id="ctl00_ContentPlaceHolder1_DropDownList_WPF"]').find_element(By.XPATH, f"//option[text()='{NAME}']").click()
    time.sleep(1)

    # 選監測項目
    # 懸浮固體 //*[@id="ctl00_ContentPlaceHolder1_RBL_CD1_0"]
    # 氨氣 //*[@id="ctl00_ContentPlaceHolder1_RBL_CD1_1"]
    # 化學需氧量 //*[@id="ctl00_ContentPlaceHolder1_RBL_CD1_2"]
    # 氫離子濃度指數 //*[@id="ctl00_ContentPlaceHolder1_RBL_CD1_3"]
    # 導電度 //*[@id="ctl00_ContentPlaceHolder1_RBL_CD1_4"]
    # 水量 //*[@id="ctl00_ContentPlaceHolder1_RBL_CD1_5"]
    # 水溫 //*[@id="ctl00_ContentPlaceHolder1_RBL_CD1_6"]
    if ITEM == "懸浮固體":
        driver.find_element(By.XPATH, '//*[@id="ctl00_ContentPlaceHolder1_RBL_CD1_0"]').click()
    elif ITEM == "氨氮":
        driver.find_element(By.XPATH, '//*[@id="ctl00_ContentPlaceHolder1_RBL_CD1_1"]').click()
    elif ITEM == "化學需氧量":
        driver.find_element(By.XPATH, '//*[@id="ctl00_ContentPlaceHolder1_RBL_CD1_2"]').click()
    elif ITEM == "氫離子濃度指數":
        driver.find_element(By.XPATH, '//*[@id="ctl00_ContentPlaceHolder1_RBL_CD1_3"]').click()
    elif ITEM == "導電度":
        driver.find_element(By.XPATH, '//*[@id="ctl00_ContentPlaceHolder1_RBL_CD1_4"]').click()
    elif ITEM == "水量":
        driver.find_element(By.XPATH, '//*[@id="ctl00_ContentPlaceHolder1_RBL_CD1_5"]').click()
    elif ITEM == "水溫":
        driver.find_element(By.XPATH, '//*[@id="ctl00_ContentPlaceHolder1_RBL_CD1_6"]').click()
    else:
        print('沒有這個監測項目!!')
    time.sleep(1)
    
    # 選日期
    # DATE = (datetime.now() - timedelta(days=1)).strftime("%Y/%m/%d")
    driver.find_element(By.XPATH, '//*[@id="ctl00_ContentPlaceHolder1_DateEdit_I"]').clear()
    driver.find_element(By.XPATH, '//*[@id="ctl00_ContentPlaceHolder1_DateEdit_I"]').send_keys(DATE)
    time.sleep(1)

    # 查詢
    driver.find_element(By.XPATH, '//*[@id="ctl00_ContentPlaceHolder1_Btn_Search"]').click()
    time.sleep(1)

    # 取得page1資料
    data1 = driver.find_elements(By.XPATH, '//*[@id="ctl00_ContentPlaceHolder1_GridView1"]')
    page1 = data1[0].text.split('\n')

    # 下一頁
    driver.find_element(By.XPATH, '//*[@id="ctl00_ContentPlaceHolder1_GridView1_DXPagerBottom_PBN"]/img').click()
    time.sleep(1)

    # 取得page2資料
    data2 = driver.find_elements(By.XPATH, '//*[@id="ctl00_ContentPlaceHolder1_GridView1"]')
    page2 = data2[0].text.split('\n')

    # 下一頁
    driver.find_element(By.XPATH, '//*[@id="ctl00_ContentPlaceHolder1_GridView1_DXPagerBottom_PBN"]/img').click()
    time.sleep(1)

    # 取得page3資料
    data3 = driver.find_elements(By.XPATH, '//*[@id="ctl00_ContentPlaceHolder1_GridView1"]')
    page3 = data3[0].text.split('\n')

    # driver.quit()
    return page1,page2,page3
```

```python
# 合併資料
def merge_data(page1,page2,page3):
    hour_list = []
    avg_list = []
    for s in page1:
        if ITEM in s:
            contents = s.split(' ')
            hour = contents[3]+":00"
            avg = contents[5]
            hour_list.append(hour)
            avg_list.append(avg)

    for s in page2:
        if ITEM in s:
            contents = s.split(' ')
            hour = contents[3]+":00"
            avg = contents[5]
            hour_list.append(hour)
            avg_list.append(avg)

    for s in page3:
        if ITEM in s:
            contents = s.split(' ')
            hour = contents[3]+":00"
            avg = contents[5]
            hour_list.append(hour)
            avg_list.append(avg)
    # 建立dataframe模板(全部時間都有的)
    time_temp = ['00:00', '01:00', '02:00', '03:00', '04:00', '05:00', '06:00', '07:00', '08:00', '09:00', '10:00', '11:00', '12:00', '13:00', '14:00', '15:00', '16:00', '17:00', '18:00', '19:00', '20:00', '21:00', '22:00', '23:00']
    # DATE = (datetime.now() - timedelta(days=1)).strftime("%Y/%m/%d")
    date_temp =[DATE for i in range(len(time_temp))]
    # avg_temp = [0 for i in range(len(time_temp))]
    data_dict = {'日期':date_temp,'時間':time_temp}
    df_temp = pd.DataFrame(data_dict)

    # 把上面抓的資料弄成dataframe
    date_list =[DATE for i in range(len(hour_list))]
    d = {'日期':date_list,
        '時間':hour_list,
        '平均值':avg_list}
    df1 = pd.DataFrame(d)

    # 兩者合併，缺失值填0
    df = pd.merge(df_temp, df1, how="outer")
    df = df.fillna(0)
    
    return df
```
